```python
#rating of the book
book_soup.find_all("p", attrs = {"class" : "star-rating"})[0]["class"][1]

#create a function that converts the rating to a number so that we can use this to sort the books
def rating_to_number(rating):
    if rating == "One":
        return 1
    elif rating == "Two":
        return 2
    elif rating == "Three":
        return 3
    elif rating == "Four":
        return 4
    elif rating == "Five":
        return 5
    else:
        return None
```

```python
def scrape_books(min_rating, max_price):
    all_books_list = []
    number_pages = int(soup.find_all("li", class_ = "current")[0].get_text().split()[3])

    for i in range(number_pages):
        page_url = f"https://books.toscrape.com/catalogue/page-{i}.html"
        page_response = requests.get(page_url)
        page_soup = BeautifulSoup(page_response.content)
        books = soup.find_all("h3")
        
        for book in books:
            book_url = url + book.find("a")["href"] # get the link to each book
            book_response = requests.get(book_url) # get the response of the link
            book_soup = BeautifulSoup(book_response.content) # get the soup of the link

            # get the info of the book and store it in a dictionary per book
            book_info = {
                "title" : book_soup.find_all("title")[0].get_text().strip().split("|")[0].strip(),
                "price" : float(book_soup.find_all("p", class_="price_color")[0].get_text().strip().replace("£", "")),
                "UPC" : book_soup.find_all("td")[0].get_text().strip(),
                "rating" : rating_to_number(book_soup.find_all("p", attrs = {"class" : "star-rating"})[0]["class"][1]),
                "genre" : book_soup.find_all("a")[3].get_text(),
                "availability" : book_soup.find_all("p", attrs = {"class" : "instock availability"})[0].get_text().strip(),
                "description" : book_soup.find_all("p")[3].get_text(),
                }
            #append the dictionary to a list of dictionaries with all the books
            #verify the conditions from the user input
            if book_info["rating"] >= min_rating and book_info["price"] <= max_price:
                all_books_list.append(book_info)

    return pd.DataFrame(all_books_list)
  
```
